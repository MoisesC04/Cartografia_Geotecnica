```python
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
cuencaRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Cuenca_tif/CuencaZona2.tif?raw=true')
cuencaB1 = cuencaRaster.read(1)

# CorreciÃ³n para organizar luego los otros valores
cuencaB1 = np.where(cuencaB1 == 1, np.nan, 1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(cuencaB1)
plt.colorbar();

print('Cuenca B1')
print('Forma: ', cuencaB1.shape)
print('Valores Ãºnicos: ', np.unique(cuencaB1))
```

```python
inventarioRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%204%20-%20Inventario/Tiff/InventarioZona.tif?raw=true')
inventarioB1 = inventarioRaster.read(1)

# CorreciÃ³n para organizar luego los otros valores
inventarioB1 = np.where(inventarioB1 == 1, inventarioB1, 0)
#inventarioB1 = np.where(inventarioB1 == 0, np.nan,  inventarioB1)
#inventarioB1 = np.where(inventarioB1 == 15, np.nan,  inventarioB1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(inventarioB1)
plt.colorbar();

print('InventarioB1')
print('Forma: ', inventarioB1.shape)
print('Valores Ãºnicos: ', np.unique(inventarioB1))
```

```python
inventarioZona = inventarioB1 * cuencaB1

#inventarioB1 = np.where(inventarioB1 == 1, inventarioB1, 0)


plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(inventarioZona)
plt.colorbar();

print('Cuenca B1')
print('Forma: ', inventarioZona.shape)
print('Valores Ãºnicos: ', np.unique(inventarioZona))
```

```python
geologiaRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Geolog%C3%ADa/GeologiaZona.tif?raw=true')
geologiaB1 = geologiaRaster.read(1)

geologiaB1 = np.where(geologiaB1 == 0, np.nan, geologiaB1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(geologiaB1)
plt.colorbar();

print('Geologia B1')
print('Forma: ', geologiaB1.shape)
print('Valores Ãºnicos: ', np.unique(geologiaB1))
```

```python
geomorfologiaRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Geomorfolog%C3%ADa/Tiff/GeomorfologiaZona.tif?raw=true')
geomorfologiaB1 = geomorfologiaRaster.read(1)

geomorfologiaB1 = np.where(geomorfologiaB1 == 0, np.nan, geomorfologiaB1)
#geomorfologiaB1 = np.where(geomorfologiaB1 == 0, geomorfologiaB1, np.nan)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(geomorfologiaB1)
plt.colorbar();

print('GeomorfologÃ­a B1')
print('Forma: ', geomorfologiaB1.shape)
print('Valores Ãºnicos: ', np.unique(geomorfologiaB1))
```

```python
pendienteRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Pendientes/Tiff/PendienteZona.tif?raw=true')
pendienteB1 = pendienteRaster.read(1)

pendienteB1 = np.where(pendienteB1 == pendienteB1.min(), np.nan, pendienteB1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(pendienteB1)
plt.colorbar();

print('Pendientes B1')
print('Forma: ', pendienteB1.shape)
print('Valores Ãºnicos: ', np.unique(pendienteB1))
```

```python
curvaturaRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Curvatura/Tiff/CurvaturaZona.tif?raw=true')
curvaturaB1 = curvaturaRaster.read(1)

curvaturaB1 = np.where(curvaturaB1 == curvaturaB1.min(), np.nan, curvaturaB1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(curvaturaB1)
plt.colorbar();

print('Curvatura B1')
print('Forma: ', curvaturaB1.shape)
print('Valores Ãºnicos: ', np.unique(curvaturaB1))
```

```python
aspectoRaster = rio.open('https://github.com/MoisesC04/Cartografia_Geotecnica/blob/main/Taller%203%20-%20Informaci%C3%B3n%20secundaria/Aspecto/Tiff/AspectoZona.tif?raw=true')
aspectoB1 = aspectoRaster.read(1)

aspectoB1 = np.where(aspectoB1 == aspectoB1.min(), np.nan, aspectoB1)

plt.figure(figsize=(40, 10))  # Ajusta el tamaÃ±o
plt.imshow(aspectoB1)
plt.colorbar();

print('Aspecto B1')
print('Forma: ', aspectoB1.shape)
print('Valores Ãºnicos: ', np.unique(aspectoB1))
```

```python
# Convirtiendo matrices a vectores con nan

inventarioVc = inventarioZona.ravel()
print('Longitud inventarioVc: ', inventarioVc.shape[0])

geologiaVc = geologiaB1.ravel()
print('Longitud geologiaVc: ', geologiaVc.shape[0])

geomorfologiaVc = geomorfologiaB1.ravel()
print('Longitud geomorfologiaVc: ', geomorfologiaVc.shape[0])

pendienteVc = pendienteB1.ravel()
print('Longitud pendienteVc: ', pendienteVc.shape[0])

curvaturaVc = curvaturaB1.ravel()
print('Longitud curvaturaVc: ', curvaturaVc.shape[0])

aspectoVc = aspectoB1.ravel()
print('Longitud aspectoVc: ', aspectoVc.shape[0])
```

```python
# Eliminando nans

inventarioVcMenM = inventarioVc[~np.isnan(inventarioVc)]
print('Longitud inventarioVcMenM: ', inventarioVcMenM.shape[0])

geologiaVcMenM = geologiaVc[~np.isnan(geologiaVc)]
print('Longitud geologiaVcMenM: ', geologiaVcMenM.shape[0])

geomorfologiaVcMenM = geomorfologiaVc[~np.isnan(geomorfologiaVc)]
print('Longitud geomorfologiaVcMenM: ', geomorfologiaVcMenM.shape[0])

pendienteVcMenM = pendienteVc[~np.isnan(pendienteVc)]
print('Longitud pendienteVcMenM: ', pendienteVcMenM.shape[0])

curvaturaVcMenM = curvaturaVc[~np.isnan(curvaturaVc)]
print('Longitud curvaturaVcMenM: ', curvaturaVcMenM.shape[0])

aspectoVcMenM = aspectoVc[~np.isnan(aspectoVc)]
print('Longitud aspectoVcMenM: ', aspectoVcMenM.shape[0])
```

```python
### Realizando el dataframe

diccionario = {'inventario':inventarioVcMenM, 'geologia':geologiaVcMenM, 'geomorfologia':geomorfologiaVcMenM, 'pendiente':pendienteVcMenM,
               'curvatura':curvaturaVcMenM, 'aspecto':aspectoVcMenM}

tablaDatos = pd.DataFrame(diccionario)
print(list(tablaDatos.columns))

tablaDatos.tail(5)
```
